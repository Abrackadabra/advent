```python
def after_t(a, t):
    name, speed, uptime, downtime = a
    full = t // (uptime + downtime)
    x = full * (speed * uptime)
    t -= full * (uptime + downtime)
    x += min(uptime, t) * speed
    return x
```

```python
time = 2503

a = []

for line in lines:
    m = re.findall(r'^(\S+) can fly (\d+) km/s for (\d+) seconds, but then must rest for (\d+) seconds.', line)
    b = array(m[0], dtype=object)
    b[1:] = vectorize(int)(b[1:])
    a.append(b)
a = array(a, dtype=object)
a
```

```python
a = array([['Comet',  14, 10, 127],
           ['Dancer', 16, 11, 162]], dtype=object)
time = 1000
a
```

```python
scores = defaultdict(int)
for passed in range(1, time + 1):
    distances = array([after_t(i, passed) for i in a])
    winners = a[:, 0][distances == max(distances)]
    for i in winners:
        scores[i] += 1
for i in a:
    print('{:>10} {:>5} {:>5}'.format(i[0], scores[i[0]], after_t(i, time)))
```

```python
for i in lines:
    m = re.findall('Sue (\d+): (.*)', i)
    num = int(m[0][0])
    parts = m[0][1].split(', ')
    d = {}
    for j in parts:
        m = re.findall(r'(\S+): (\d+)', j)
        d[m[0][0]] = int(m[0][1])
    ok = True
    for j, k in d.items():
        if j in waldo_eq:
            ok &= waldo_eq[j] == k
    for j, k in d.items():
        if j in waldo_gr:
            ok &= waldo_gr[j] < k
    for j, k in d.items():
        if j in waldo_le:
            ok &= waldo_le[j] > k
    if ok:
        print(num)
```

```python
a = read('18a.input')

a = array([[i == '#' for i in j] for j in a])
```

```python
def neighbours(x, y, n, m):
    for i, j in itertools.product(range(-1, 2), range(-1, 2)):
        if 0 <= x + i < n and 0 <= y + j < m and (i, j) != (0, 0):
            yield x + i, y + j

def life_step(a):
    n, m = a.shape
    b = zeros((n, m), dtype=int)
    for x, y in itertools.product(range(n), range(m)):
        s = sum(a[i] for i in neighbours(x, y, n, m))
        b[x, y] = s == 3 or (s == 2 and a[x, y])
    return b

t = a
for i in range(100):
    t = life_step(t)
print(sum(t))
```

```python
corners = array(list(itertools.product([-1, 0], [-1, 0])))

def life_step(a):
    n, m = a.shape
    a[corners[:, 0], corners[:, 1]] = 1
    b = zeros((n, m), dtype=int)
    for x, y in itertools.product(range(n), range(m)):
        s = sum(a[i] for i in neighbours(x, y, n, m))
        b[x, y] = s == 3 or (s == 2 and a[x, y])
    b[corners[:, 0], corners[:, 1]] = 1
    return b

t = a
for i in range(100):
    t = life_step(t)

print(sum(t))
```
